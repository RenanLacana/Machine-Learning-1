```python
import pandas as pd 

uri='https://gist.githubusercontent.com/guilhermesilveira/2d2efa37d66b6c84a722ea627a897ced/raw/10968b997d885cbded1c92938c7a9912ba41c615/tracking.csv'

dados = pd.read_csv(uri)
dados.head()
```

```python
mapa = {
    'home' : 'pagina_principal',
    'how_it_works' : 'como_funciona',
    'contact' : 'contato',
    'bought' : 'comprou'
}
```

```python
dados.rename(mapa, axis=1, inplace=True)
```

```python
dados.head()
```

```python
x = dados[['pagina_principal', 'como_funciona', 'contato']]
y = dados['comprou']
```

# Primeiramente vamos dividir nossos dados em grupos de treino e teste para nosso modelo. Nessa parte faremos de uma forma manual.

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
```

```python
dados.shape
```

### Treinaremos o modelo de classificação com 75 elementos e testaremos com 24 elementos.

```python
x_treino = x[:75]
x_teste = x[75:]
```

```python
y_treino = y[:75]
y_teste = y[75:]
```

### Instanciando modelo de Machine Learning

```python
modelo = LinearSVC()
modelo.fit(x_treino, y_treino)
y_previsto = modelo.predict(x_teste)
acuracia = accuracy_score(y_teste, y_previsto) * 100
```

```python
print('A acurácia do modelo foi de %.2f'% acuracia + " %")
```

# Agora vamos refatorar o que fizemos acima incluindo algumas bibliotecas e variáveis que nos auxiliam.

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

seed = 5
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, random_state = seed, test_size=0.25, stratify=y)
```

```python
modelo = LinearSVC()
modelo.fit(x_treino, y_treino)
y_previsto = modelo.predict(x_teste)
acuracia = accuracy_score(y_teste, y_previsto) * 100
```

```python
print('A acurácia do modelo foi de %.2f'% acuracia + " %")
```
